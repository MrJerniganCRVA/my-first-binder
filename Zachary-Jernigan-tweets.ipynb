```python
import numpy as np
import pandas as pd
import json
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import subjectivity
from nltk.sentiment import SentimentAnalyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from nltk.sentiment.util import *
from nltk.corpus import stopwords
```

```python
#need to download the important information from nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('universal_tagset')
nltk.download('maxent_ne_chunker')
nltk.download('words') 
nltk.download('treebank')
nltk.download('stopwords')
```

<h2>Importing Data from JSON File</h2>

```python
with open('bluesky-data.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

df = pd.DataFrame(data)
df.head() #To check to see if it loaded properly
```

```python
#Removing Stopwords
sws = set(stopwords.words('english'))
text_nosws = []

for j,text in enumerate(df['text']):
    text_nosws.append(' '.join([i for i in text.lower().split() if i not in sws]))

df["Cleaned Text"] = text_nosws
df.head() #Just checking I have what I think I have
```

<h2>Sentiment Analysis</h2>

```python
#now going to add sentiment analysis to the data frame...wish me luck

comp = []
neg = []
neu = []
pos = []

for text in text_nosws:
    sid = SentimentIntensityAnalyzer()
    ss = sid.polarity_scores(text)
    comp.append(ss['compound'])
    neg.append(ss['neg'])
    neu.append(ss['neu'])
    pos.append(ss['pos'])

df['Compound Score'] = comp
df['Negative Score'] = neg
df['Neutral Score'] = neu
df['Positive Score'] = pos

df.head()
    
```

```python
#Making Top 10 sorted lists for analysis
df_sorted_compound = df.sort_values(by='Compound Score', ascending=False)[:5]
df_sorted_neg = df.sort_values(by='Negative Score', ascending=False)[:5]
df_sorted_neu = df.sort_values(by='Neutral Score', ascending=False)[:5]
df_sorted_pos = df.sort_values(by='Positive Score', ascending=False)[:5]
```

<h4>Showing top 5 posts by Compound, Negative, Neutral, and Positive Score</h4>
